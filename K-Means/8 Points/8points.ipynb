### Question
Clustering: K-Means <br>
We have given a collection of 8 points. P1=[0.1,0.6], P2=[0.15,0.71], P3=[0.08,0.9], P4=[0.16, 0.85], P5=[0.2,0.3], P6=[0.25,0.5], P7=[0.24,0.1] & P8=[0.3,0.2]. <br>
Perform the k-mean clustering with initial centroids as m1=P1 =Cluster#1=C1 and m2=P8=cluster#2=C2.  <br>
Answer the following:
1. Which cluster does P6 belongs to? 
2. What is the population of a cluster around m2?
3. What is the updated value of m1 and m2?

```python
# Import necessary libraries
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Given points
points = np.array([[0.1, 0.6], [0.15, 0.71], [0.08, 0.9], [0.16, 0.85],
                   [0.2, 0.3], [0.25, 0.5], [0.24, 0.1], [0.3, 0.2]])

# Initial centroids
m1 = np.array([0.1, 0.6])
m2 = np.array([0.3, 0.2])
```

```python
# Perform K-Means clustering
kmeans = KMeans(n_clusters=2, init=np.array([m1, m2]), n_init=1, random_state=42)
clusters = kmeans.fit_predict(points)
```

```python
# Visualize the clusters
plt.scatter(points[:, 0], points[:, 1], c=clusters, cmap='viridis', marker='o')
plt.scatter(m1[0], m1[1], c='red', marker='x', label='m1 (Initial)')
plt.scatter(m2[0], m2[1], c='blue', marker='x', label='m2 (Initial)')
plt.title('K-Means Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
# Answers to the questions
p6_cluster = kmeans.predict([[0.25, 0.5]])[0] + 1  # Adding 1 to make clusters 1-indexed
m2_population = np.sum(clusters == 1)

m1_updated = kmeans.cluster_centers_[0]
m2_updated = kmeans.cluster_centers_[1]

print(f"1] P6 belongs to Cluster {p6_cluster}")
print(f"2] Population of Cluster around m2: {m2_population}")
print(f"3] Updated value of m1: [{m1_updated[0]:.3f}, {m1_updated[1]:.3f}], Updated value of m2: [{m2_updated[0]:.3f}, {m2_updated[1]:.3f}]")
```
