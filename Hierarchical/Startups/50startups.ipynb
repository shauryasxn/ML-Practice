Apply Hierarchical clustering algorithms (based on PROFIT column) to find the group of start-ups.

```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('50_Startups.csv')
```

```python
# Extract the feature for clustering
X = df[['PROFIT']]

# Standardize the data
X_scaled = StandardScaler().fit_transform(X)
```

```python
# Apply Hierarchical Clustering
cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
df['cluster'] = cluster.fit_predict(X_scaled)
```

```python
# Visualize the clusters
sns.scatterplot(x='PROFIT', y='RND', hue='cluster', data=df)
plt.title('Hierarchical Clustering - Profit vs. R&D Spending')
plt.xlabel('Profit')
plt.ylabel('R&D Spending')
plt.show()
```
