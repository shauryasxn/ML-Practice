Also Display the regression results and plot the regression linea for all three.

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('advertising.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
# Explore whether TV advertising spending can predict the number of sales
X_tv = df[['TV']]
y_tv = df['Sales']

# Split the data into training and testing sets
X_tv_train, X_tv_test, y_tv_train, y_tv_test = train_test_split(X_tv, y_tv, test_size=0.2, random_state=42)

# Create a linear regression model for TV
model_tv = LinearRegression()
model_tv.fit(X_tv_train, y_tv_train)
y_tv_pred = model_tv.predict(X_tv_test)

# Plot the regression line for TV
plt.scatter(X_tv_test, y_tv_test, color='black', label='Actual Data')
plt.plot(X_tv_test, y_tv_pred, color='blue', linewidth=3, label='Regression Line')
plt.title('Linear Regression - TV vs. Sales')
plt.xlabel('TV Advertising Spending')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Display the regression results for TV
r_squared_tv = model_tv.score(X_tv_test, y_tv_test)
mse_tv = mean_squared_error(y_tv_test, y_tv_pred)
mae_tv = mean_absolute_error(y_tv_test, y_tv_pred)
print(f'TV - Coefficient of Determination (R-squared): {r_squared_tv:.4f}')
print(f'TV - Mean Squared Error (MSE): {mse_tv:.4f}')
print(f'TV - Mean Absolute Error (MAE): {mae_tv:.4f}')
```

```python
# Explore whether Radio advertising spending can predict the number of sales
X_radio = df[['Radio']]
y_radio = df['Sales']

# Split the data into training and testing sets
X_radio_train, X_radio_test, y_radio_train, y_radio_test = train_test_split(X_radio, y_radio, test_size=0.2, random_state=42)

# Create a linear regression model for Radio
model_radio = LinearRegression()
model_radio.fit(X_radio_train, y_radio_train)
y_radio_pred = model_radio.predict(X_radio_test)

# Plot the regression line for Radio
plt.scatter(X_radio_test, y_radio_test, color='black', label='Actual Data')
plt.plot(X_radio_test, y_radio_pred, color='blue', linewidth=3, label='Regression Line')
plt.title('Linear Regression - Radio vs. Sales')
plt.xlabel('Radio Advertising Spending')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Display the regression results for Radio
r_squared_radio = model_radio.score(X_radio_test, y_radio_test)
mse_radio = mean_squared_error(y_radio_test, y_radio_pred)
mae_radio = mean_absolute_error(y_radio_test, y_radio_pred)
print(f'Radio - Coefficient of Determination (R-squared): {r_squared_radio:.4f}')
print(f'Radio - Mean Squared Error (MSE): {mse_radio:.4f}')
print(f'Radio - Mean Absolute Error (MAE): {mae_radio:.4f}')
```
